```python
from pymongo import MongoClient
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bbc.com/news'
```

```python
req = requests.get(url)
html = req.text

print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print("Total articles found:", len(soup.find_all('article')))
print("First article:", soup.find('article'))
```

```python
articles = soup.find_all('article')
for i, art in enumerate(articles[:5]):
    print(f"Article {i}: class={art.get('class')}, data-testid={art.get('data-testid')}")
```

```python
h1s = soup.find_all('h1')
print("H1s found:", len(h1s))
for h in h1s[:5]:
    print(h.text)
```

```python
h3s = soup.find_all('h3')
print("H3s found:", len(h3s))
for h in h3s[:5]:
    print(h.text)
```

```python
items = soup.find_all('article', attrs={'class' : 'set'})
```

```python
for item in items:
    print(item.find('h1').text)
```

```python
kits_info = []
for item in soup.find_all('h3'):
    titulo = item.text.strip()
    parent_a = item.find_parent('a')
    link = parent_a['href'] if parent_a else ''
    kit = {
        'titulo': titulo,
        'link': link
    }
    kits_info.append(kit)

display(kits_info)
```

```python
client = MongoClient('mongodb://root:rootpass@localhost:27017')
client.server_info()['version']
```

```python
db = client['bbc']
collection = db['kits']
```

```python
collection.insert_many(kits_info)
```

```python
for doc in collection.find():
    display(doc)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('kits.db')
```

```python
cursor = conn.cursor();

cursor.execute("""
    CREATE TABLE IF NOT EXISTS kits(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        titulo TEXT NOT NULL,
        link TEXT NOT NULL
    )
""")
```

```python
for kit in kits_info:
    cursor.execute(
        "INSERT INTO kits(titulo, link) VALUES (?, ?)",
        (kit['titulo'], kit['link'])
    )

conn.commit()
```

```python
cursor.execute("SELECT * FROM kits")
for item in cursor.fetchall():
    print(item)

conn.close()
```
